<p style="font-family:Helvetica;color:#ffff00">Analysis of correlation between Features and Target Variable</p>

```python
# Compute the correlation matrix
'''
    Correlation Coefficient:
        The correlation coeficient is used to analyze the relationship between two Variables, to measure their degree dependency, inorder to predict
        the behaviour or the value of One dependent Variable, in terms of Behaviour or Change in Value of the Other Dependent/Independent Variable.
    Correlation Matrix:
        The Correlation matrix is used to analyze the Correlation of one Feature, with all the Other Features in a Data Set.
        The Heatmap of the Correlation matrix, provides insight about the Correlation between Variables/Features in the Dataset.   
'''
correlation_matrix = Data_Frame.corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Correlation_Thresholds = [0.3,0.4,0.5,0.6,0.7]

# Data_Frame = Data_Frame.drop(["MEDV"],axis=1) -> if the Target Variable Column, is removed here, then the Correlation analysis cannot be performed further.

# Get the features to eliminate (those with correlation less than the threshold)
# The below Dict is used to Store all the features which are low than the corresponding Correlation Threshold, for feature Elimination.
Low_Correlation_Features = {}
for i in Correlation_Thresholds:
    Low_Correlation_Features.update({i:Data_Frame.columns[Data_Frame.corr()["MEDV"].abs() < i]})
High_Equal_Correlation_Features = {}
for i in Correlation_Thresholds:
    High_Equal_Correlation_Features.update({i:Data_Frame.columns[Data_Frame.corr()["MEDV"].abs() >= i]})

# Drop the low-correlation features from the DataFrame, and create a copy of the altered Dataframe for Analysis.
# print(Low_Correlation_Features,"\n\n")
# print(High_Equal_Correlation_Features)

print("Actual Data Frame's Features:")
print(Data_Frame.columns,"\n")
for i in range(len(Correlation_Thresholds)):
    print("\nFeatures(Columns) in the Data Set with Correlation with the Target Variable 'MEDV' above and equal to the Correlation Threshold ->",Correlation_Thresholds[i],":")
    print(High_Equal_Correlation_Features[Correlation_Thresholds[i]])
```

```python
Selected_Features = High_Equal_Correlation_Features[0.4]
# Inter-Quartiles Analysis:
Q_025 = []
Q_05  = []
Q_075 = []
IQR=[]
for i in range(len(Selected_Features)):
    Q_025.append(Data_Frame[Selected_Features[i]].quantile(0.25))
    Q_05.append(Data_Frame[Selected_Features[i]].quantile(0.5))
    Q_075.append(Data_Frame[Selected_Features[i]].quantile(0.75))
    IQR.append(Q_075[i]-Q_025[i])
    
for i in range(len(Selected_Features)):
    print("Inter-Quartiles of Feature",Selected_Features[i],"in Data Set")
    print("0.25 Q :",Q_025[i],"\n0.5 Q  :",Q_05[i],"\n0.75 Q :",Q_075[i],"\nIQR :",IQR[i])
```
